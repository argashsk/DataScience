```python
df_mod = pd.get_dummies(df)
dfd_mod = pd.get_dummies(dfd)
```

```python
# ambil set kolom dari masing-masing dataframe
set3 = set(df_mod.columns)
set4 = set(dfd_mod.columns)

# ambil irisan dari kedua set kolom
intersection = set3.intersection(set4)

# buat dataframe baru hanya dengan kolom yang terdapat pada irisan
df11 = df_mod[intersection]
df22 = dfd_mod[intersection]
```

```python
y_train = df11["SalePrice"].values
X_train = df11.drop(["SalePrice"],axis=1)
X_test = df22.drop(["SalePrice"],axis=1)
```

```python
lr = RandomForestRegressor()
model_lr = lr.fit(X_train,y_train)
y_pred = model_lr.predict(X_test)
```

```python
y_prediction = np.round(y_pred).flatten()
```
